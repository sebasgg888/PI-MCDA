## Entendimiento del Negocio

Con esta investigación se pretende establecer un marco de trabajo para tratar distintas variables de un modelo de recursos minerales de una veta de oro-plata con el objetivo de agrupar bloques con características similares y luego clasificarlos en categorías en un enfoque consistente, reproducible y auditable. El método busca integrar datos cualitativos, como las varianzas de estimación, números de compositos usados en la interpolación, distancia promedio a los compositos, pasadas de estimación, la varianza local del bloque, entre otras. El resultado será una clasificación consistente en categorías de recursos comparable a una clasificación hecha manualmente por un QP o CP, pero generada bajo un enfoque automatizado, con una sustancial reducción del tiempo para la ejecución de esta tarea, y el menor grado de subjetividad posible.

El set de datos en este estudio es un modelo de bloques 3D estimado en un yacimiento subterráneo de oro-plata en el departamento de Antioquia - Colombia.

### Descripción de variables

**PROPIEDADES DEL BLOQUE:** <br>
**1. XC =** Coordenada X del centroide del bloque (anonimizada)<br>
**2. YC =** Coordenada Y del centroide del bloque (anonimizada)<br>
**3. ZC =** Coordenada Z del centroide del bloque (anonimizada)<br>
**4. XINC =** Dimensión en X del bloque (m)<br>
**5. YINC =** Dimensión en Y del bloque (m)<br>
**6. ZINC =** Dimensión en Z del bloque (m)<br>
**7. Au_ppm =** Tenor final de oro estimado en partes por millón<br>
**8. Ag_ppm =** Tenor final de cobre estimado en partes por millón<br>
**9. AuE =** Relación de oro equivalente = Au_ppm + (Ag_ppm / 80)<br>
**10. Zn_ppm =** Tenor final de zinc estimado en partes por millón<br>
**11. Pb_ppm =** Tenor final de plomo estimado en partes por millón<br>
**12. Cu_ppm =** Tenor final de cobre estimado en partes por millón<br>

**PARÁMETROS DE ESTIMACIÓN:** <br>
**1. NSAMP =** Mínimo número de muestras usadas para estimar el bloque<br>
**2. NHOLES =** Mínimo número de pozos usados para la estimación del bloque<br>
**3. PASS =** Referencia al volumen de búsqueda (rango desde 1 a 5)<br>
**4. DISTNEAR =** Distancia transformada de la muestra más cercana al bloque<br>
**5. ESTVAR =** Varianza kriging de estimación (Kriging Ordinario)<br>
**6. BE =** Error de estimación del bloque = (ESTVAR*1.96)/(Au_ppm*sqrt(NSAMP))<br>

**CLASIFICACIÓN DEL QP (Empresa):** <br>
**1. CLASS_QP =** Clasificación final de la persona calificada ó QP: (1) = Medido; (2) = Indicado; (3) = Inferido.

### Matriz de correlación lineal (pearson)

El análisis de dependencia de los predictores al igual que en regresión, también es importante en problemas de clasificación; ya que, estas pueden introducir ruido en los modelos de clasificación.

```python
# Matriz de Correlación lineal de features (pearson)
#====================================================================================
hm = sns.heatmap(X.corr(method = 'pearson'), annot = True, square = False, cmap="rainbow")
hm.set(xlabel='\nbmv100 Estimation Parameters', ylabel='\nbmv100 Estimation Parameters\t', 
       title = "Matriz de Correlación de bmv100 Model (pearson)\n")
plt.show()
```

```python
# Encontrar variables correlacionadas (umbral = 0.8)
#====================================================================================
def correlation(dataset, threshold):
    col_corr = set()  
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # el abs del coeficiente
                colname = corr_matrix.columns[i]  # nombre de la columna
                col_corr.add(colname)
    return col_corr

corr_features = correlation(X, 0.8)
print('Variables Correlacionadas: ', len(set(corr_features))); print(corr_features)
```

Vemos que un par de variables se correlacionan fuertemente: NSAMP y NHOLES.

### Matriz de correlación y scatterplot (pearson)

```python
# Matriz de correlación y scatter plots
# Cortesia: https://stackoverflow.com/questions/48139899/
#====================================================================================
%%time
X_sub = X.loc[np.random.choice(X.index, 10000, replace = False)]  # subset por limitación de computo

def corrdot(*args, **kwargs):
    corr_r = args[0].corr(args[1], 'pearson')
    corr_text = round(corr_r, 2)
    ax = plt.gca()
    font_size = abs(corr_r) * 80 + 5
    ax.annotate(corr_text, [.5, .5,],  xycoords="axes fraction",
                ha='center', va='center', fontsize=font_size)

def corrfunc(x, y, **kws):
    r, p = stats.pearsonr(x, y)
    p_stars = ''
    if p <= 0.05:
        p_stars = '*'
    if p <= 0.01:
        p_stars = '**'
    if p <= 0.001:
        p_stars = '***'
    ax = plt.gca()
    ax.annotate(p_stars, xy=(0.65, 0.6), xycoords=ax.transAxes,
                color='red', fontsize=70)
    
sns.set(style='white', font_scale=1.3)
g = sns.PairGrid(X_sub, aspect=1.5, diag_sharey=False, despine=False)
g.map_lower(sns.regplot, lowess=True, ci=False,
            line_kws={'color': 'red', 'lw': 1},
            scatter_kws={'color': 'black', 's': 20})
g.map_diag(sns.distplot, color='black',
           kde_kws={'color': 'red', 'cut': 0.7, 'lw': 1},
           hist_kws={'histtype': 'bar', 'lw': 2,
                     'edgecolor': 'k', 'facecolor':'grey'})
g.map_diag(sns.rugplot, color='black')
g.map_upper(corrdot)
g.map_upper(corrfunc)
g.fig.subplots_adjust(wspace=0.1, hspace=0.1)

# Remove axis labels
for ax in g.axes.flatten():
    ax.set_ylabel('')
    ax.set_xlabel('')

# Add titles to the diagonal axes/subplots
for ax, col in zip(np.diag(g.axes), X_sub.columns):
    ax.set_title(col, y=0.82, fontsize=20)
```

El análisis de dependencia refleja una fuerte dependencia lineal entre NSAMP y NHOLES (número de muestras usadas en la estimación del bloque y el número de pozos). Lo cual tiene su explicación en el hecho que las muestras provienen de los pozos de perforación (NHOLES = NSAMP/3). Se concluye entonces que este par de variables son redundantes para los modelos de clasificación.

### Revisión del número de observaciones de cada clase

Es importante revisar el número de observaciones para cada clase. Si bien en problemas de clasificación es recomendable tener un buen balance entre clases, no se recomienda realizar balanceo de clases en los casos que exista una cantidad representativa de la clase minoritaria (en nuestro caso 292.364 observaciones de clase 1); adicionalmente algo de sesgo puede ser introducido en casos de submuestreo. La recomendación es que si el desbalanceo no es severo (como nuestro caso: 5:7:8), no es necesario balancear. Otro aspecto es que el entrenamiento en datos balanceados o equilibrados puede no ser generalizable a datos del mundo real que están naturalmente desequilibrados.
